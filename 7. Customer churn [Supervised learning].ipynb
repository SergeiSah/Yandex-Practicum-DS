### Подготовка данных

Вначале обучим модели без балансировки данных. Перед этим проведем заполнение пропусков, нормировку и кодирование признаков.

```python
# удалим записи с пропусками
churn_df = churn_df.dropna()
print('Размер датасета после удаления пропусков:', churn_df.shape)
```

Для обучения не будем использовать колонки `RowNumber`, `CustomerId` и `Surname`, поскольку они не несут в себе "полезной нагрузки". `RowNumber` по сути просто индекс строки, у него все значения уникальны. Также и для `CustomerId` - все значения уникальны, из них нельзя выделить какие-либо закономерности. Возможно `Surname` как-то и помогло бы, но на мой взгляд это как предсказывать всю судьбу человека только по тому, как его назвали, поэтому этот признак тоже не используем.

```python
df_dropped = churn_df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
# разделим выборку на обучающую, валидационную и тестовую в соотношении 3:1:1
X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(
    df_dropped, target='Exited', train_size=0.6, valid_size=0.2, test_size=0.2, random_state=42
)

print('Train size:', X_train.shape[0])
print('Valid size:', X_valid.shape[0])
print('Test size:', X_test.shape[0])
```

```python
# проведем кодировку категориальных признаков
X_ohe_train = pd.get_dummies(X_train, drop_first=True)
X_ohe_valid = pd.get_dummies(X_valid, drop_first=True)
X_ohe_test = pd.get_dummies(X_test, drop_first=True)

# для контроля выведем первые пять строк обучающей выборки
X_ohe_train.head()
```

```python
# проведем нормировку численных признаков
# чтобы не допустить утечку данных, обучим scaler только на обучающих выборках, а затем применим к остальным
cols_to_scale = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary', 'Tenure', 'NumOfProducts']

pd.options.mode.chained_assignment = None
scaler = StandardScaler()
scaler.fit(X_ohe_train[cols_to_scale]);
```

```python
# применим scaler к обучающей выборке
X_ohe_train[cols_to_scale] = scaler.transform(X_ohe_train[cols_to_scale])
display(X_ohe_train.head())
```

```python
# к валидационной
X_ohe_valid[cols_to_scale] = scaler.transform(X_ohe_valid[cols_to_scale])
display(X_ohe_valid.head())
```

```python
# к тестовой
X_ohe_test[cols_to_scale] = scaler.transform(X_ohe_test[cols_to_scale])
display(X_ohe_test.head())
```

```python
# Поскольку данные X_ и X_ohe отличаются только категориальными признаками, используем тот же scaler на оставшиеся данные
X_train[cols_to_scale] = scaler.transform(X_train[cols_to_scale])
X_valid[cols_to_scale] = scaler.transform(X_valid[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

### Обучение моделей

На первом этапе обучаем модели без балансировки данных по целевой переменной. Подбирать оптимальные гиперпараметры моделей мы будем с использованием `OptunaSearchCV`, которая также проводит кросс-валидацию.

```python
# создадим функцию для построения кривых ROC

def plot_roc_auc(y_train, y_train_proba, y_valid, y_valid_proba, valid='valid'):
    # рассчитываем показатель ROC-AUC
    train_auc = roc_auc_score(y_train, y_train_proba)
    valid_auc = roc_auc_score(y_valid, y_valid_proba)
    
    # строим график
    plt.figure(figsize=(7, 5))
    
    plt.plot(np.linspace(0,1,100), np.linspace(0,1,100), linestyle='--')
    plt.plot(*roc_curve(y_train, y_train_proba)[:2], label=f'train AUC={train_auc:.4f}')
    plt.plot(*roc_curve(y_valid, y_valid_proba)[:2], label=f'{valid} AUC={valid_auc:.4f}')
    
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-curve')
    
    plt.show()
```

```python
# создадим функцию, в которую можно будет передавать обучающие данные и модель с сеткой гиперпараметров
# которая будет искать их оптимальную комбинацию для моделей DecisionTreeClassifier, RandomForest и LogisticRegression
# дополнительно будем строить ROC-curve и сравнивать полученный score на обучающих данных с валидационными

def search_optimal_hyperparameters(X_train, y_train, X_valid, y_valid, 
                                   model, grid_parameters, 
                                   cv=4, 
                                   scoring='f1',
                                   print_info=True,
                                   plot=True
                                  ):
    
    # используем grid search из модуля optuna
    cv_model = OptunaSearchCV(
        estimator=model,
        param_distributions=grid_parameters,
        cv=cv,
        scoring=scoring,  # метрика, которую необходимо улучшить
        verbose=0,
        n_jobs=-1,        # используем все доступные ядра процессора
        random_state=101,
        refit=True        # выбираем модель с наилучшими параметрами
    )
    
    cv_model.fit(X_train, y_train)
    
    if print_info:
        # Выведем получившиеся оптимальные параметры
        print('Optimal parameters')
        for param, val in cv_model.best_estimator_.get_params().items():
            if param in grid_parameters:
                print(f'    {param}: {val}')
                
        print('\n    Значения метрик на обучающей выборке')
        print(classification_report(y_train, cv_model.predict(X_train)))
                
        print('    Значения метрик на валидационной выборке')
        print(classification_report(y_valid, cv_model.predict(X_valid)))
    
    if plot:
        # строим кривую ROC
        plot_roc_auc(y_train, cv_model.predict_proba(X_train)[:, 1],
                     y_valid, cv_model.predict_proba(X_valid)[:, 1])
    
    return cv_model
```

```python
%%time

# задаем сетку с различными значениями гиперпараметров решающего дерева
params_dtree = {
    'criterion': CategoricalDistribution(['gini', 'entropy']), # функция определения качества разбиения подвыборки
    'max_depth': IntDistribution(1, 16, step=1),               # максимальная глубина решающего дерева
    'max_leaf_nodes': IntDistribution(2, 20, step=1),          # максимальное число листьев в дереве
    'min_samples_leaf': IntDistribution(1, 10, step=1),        # минимальное число объектов в листе
    'min_samples_split': IntDistribution(2, 11), # минимальное количество элементов в вершине, чтобы её можно было делить
    'class_weight': CategoricalDistribution(['balanced'])
}

dtree = search_optimal_hyperparameters(X_ohe_train, y_train, X_ohe_valid, y_valid,
                                          DecisionTreeClassifier(random_state=101), params_dtree)
```

### Upsampling

```python
# создадим функцию, которая увеличивает число записей с меньшим классом

def upsample(X, y):
    df = pd.concat([X, y], axis=1)
    df_majority = df[df.Exited == False]
    df_minority = df[df.Exited == True]
    
    # увеличиваем число строк в датафрейме с меньшим числом записей за счет случайного дублирования
    df_minority_upsampled = resample(df_minority, 
                                     n_samples=df_majority.shape[0],
                                     random_state=42)
    
    # собираем всё в единый датафрейм, перемешиваем строки
    df_upsampled = shuffle(pd.concat([df_majority, df_minority_upsampled])) 
    
    return df_upsampled.drop('Exited', axis=1), df_upsampled.Exited
```

### Downsampling

```python
# создадим функцию, которая уменьшает число записей с меньшим классом

def downsample(X, y, frac=0.2, rs=42):
    df = pd.concat([X, y], axis=1)
    df_majority = df[df.Exited == False]
    df_minority = df[df.Exited == True]
    
    df_majority_downsampled = df_majority.sample(frac=frac, random_state=rs)
    df_downsampled = shuffle(pd.concat([df_majority_downsampled, df_minority]), random_state=rs)
    
    return df_downsampled.drop('Exited', axis=1), df_downsampled.Exited
```

```python
# зададим такое соотношение, чтобы число записей в классах было равным
frac = X_ohe_train[y_train == True].shape[0] / X_ohe_train[y_train == False].shape[0]
print('Fraction:', frac)

X_down_ohe_train, y_down_ohe_train = downsample(X_ohe_train, y_train, frac=frac)
X_down_train, y_down_train = downsample(X_train, y_train, frac=frac)
```

### Выбор оптимального порога

Для уже обученных на разных выборках моделей, определим оптимальные пороги.

```python
# функция для определения оптимального порога, которая возвращает также максимальное значение f1

def best_f1(y, y_predict):
    f1_list = []
    
    for threshold in np.linspace(0, 1, 101):
        f1_list.append(f1_score(y, y_predict > threshold))
        
    f1_max = max(f1_list)
    threshold_max = f1_list.index(f1_max) / 100
    
    return f1_max, threshold_max
```

```python
# зададим функцию для построения кривых Precision-Recall

def plot_PR(y_train, y_train_proba, y_valid, y_valid_proba, title='PR-curve'):
    plt.figure(figsize=(7, 5))
    
    # рассчитываем значения метрик Precision и Recall для разных порогов
    precision_train, recall_train, thresholds_train = precision_recall_curve(y_train, y_train_proba)
    precision_valid, recall_valid, thresholds_valid = precision_recall_curve(y_valid, y_valid_proba)
    
    # определяем пороги, при которых достигается максимум f1
    f1_train_best, threshold_train_best = best_f1(y_train, y_train_proba)
    f1_valid_best, threshold_valid_best = best_f1(y_valid, y_valid_proba)
   
    
    # строим кривые, задаем подписи
    plt.step(recall_train, precision_train, label=f'Train: best f1={f1_train_best:.2f}, threshold={threshold_train_best:.2f}')
    plt.step(recall_valid, precision_valid, label=f'Valid: best f1={f1_valid_best:.2f}, threshold={threshold_valid_best:.2f}')
    
    legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")
    
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(title, fontsize=16)
    
    plt.show()
    
    # будем возвращать оптимальный f1 и значение порога для него для валидационной выборки
    return f1_valid_best, threshold_valid_best
```

```python
# создадим функцию для вывода основных метрик модели и построения кривых ROC
def test_model(y_train, predicted_train, y_test, predicted_test, model, threshold, plot_features=True):
    # выводим основные метрики классификации модели
    print('Выбранный порог:', threshold)
    print('\n    Значения метрик на тестовой выборке')
    print(classification_report(y_test, predicted_test > threshold))
    
    # строим матрицу confusion matrix
    cm = pd.DataFrame(confusion_matrix(y_test, predicted_test > threshold), 
                      index=[True, False], columns=[True, False])
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title('Confusion matrix', fontdict={'size':16})
    
    # проверяем модель на вменяемость
    accuracy = accuracy_score(y_test, predicted_test > threshold)
    if accuracy > accuracy_threshold:
        print(f'Accuracy {accuracy:.2f} - модель вменяема')
    else:
        print(f'Accuracy {accuracy:.2f} - модель невменяема')
    
    # строим ROC-AUC
    plot_roc_auc(y_train, predicted_train, y_test, predicted_test, valid='test')
    
    if plot_features:
        # выведем наиболее важные параметры
        important_X = pd.DataFrame({'X': model.best_estimator_.feature_importances_}, 
                                   index=model.best_estimator_.feature_names_in_).sort_values(by='X', ascending=False)
        important_X.plot(kind='bar', title='Important features', figsize=(7,5))
    
    return f1_score(y_test, predicted_test > threshold), accuracy
```

#### Решающее дерево

```python
# здесь будем хранить значения метрик протестированных моделей
f1_list = []
accuracy_list = []

# выбираем модель с наибольшим значением f1
ind = best_dtree_thresholds.index(max(best_dtree_thresholds))
model = [dtree, dtree_up, dtree_down][ind]

f1, acc = test_model(y_train, model.predict_proba(X_ohe_train)[:, 1], y_test, model.predict_proba(X_ohe_test)[:, 1],
                     model, best_dtree_thresholds[ind][1])
f1_list.append(f1)
accuracy_list.append(acc)
```
